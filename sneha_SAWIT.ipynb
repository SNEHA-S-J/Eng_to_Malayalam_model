```python
from datasets import load_dataset
from transformers import MarianTokenizer

# Load dataset
dataset = load_dataset("SnehaJay7/malayalam")

# Load tokenizer
model_checkpoint = "Helsinki-NLP/opus-mt-en-ml"
tokenizer = MarianTokenizer.from_pretrained(model_checkpoint)

# Tokenization function
def tokenize_function(example):
    return tokenizer(example["english"], text_target=example["malayalam"], padding="max_length", truncation=True)

# Apply tokenization
tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
# Split dataset into train and validation sets
split_dataset = tokenized_datasets["train"].train_test_split(test_size=0.2)
train_dataset = split_dataset["train"]
eval_dataset = split_dataset["test"]
```

```python
train_dataset = train_dataset.shuffle(seed=42).select(range(1000))  # Train on only 1000 samples
eval_dataset = eval_dataset.shuffle(seed=42).select(range(200))  # Evaluate on only 200 samples
```
